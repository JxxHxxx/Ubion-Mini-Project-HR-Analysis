```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
plt.rcParams['font.family'] = 'malgun gothic'
```

```python
HR = pd.read_csv('HR_comma_sep.csv')
```

```python
HR.columns[0]
```

```python
j = 0 
for i in HR.columns:
    print('-----------------------------featuresName--%s----------------------------------'%(HR.columns[j]))
    print(HR[i].unique())
    j += 1
```

```python
plt.title('임금 그룹')
HR['salary'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.title('진행중인 프로젝트 갯수')
HR['number_project'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.title('부서별 인구')
HR['sales'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.title('사고 발생 유무')
HR['Work_accident'].value_counts().plot(kind='bar')
plt.show()
```

```python
grouped = HR['Work_accident'].groupby(HR['sales'])
```

```python
plt.title('부서 별 사고율')
(grouped.sum()/grouped.size()).plot(kind='bar')
plt.show()
```

```python
grouped_2 = HR['time_spend_company'].groupby(HR['sales'])
```

```python
plt.title('부서 별 근무 시간')
(grouped_2.sum()/grouped_2.size()).plot(kind='bar')
plt.show()
```
